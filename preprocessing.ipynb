```python
import pandas as pd
from datetime import datetime
import json
```

```python
# Load the dataset
data = pd.read_csv('./data/kc_house_data.csv')
```

```python
data.head(5)
```

```python
# Convert date into timestamp
#data["timestamp"] = data.apply(lambda x: datetime.strptime(x['date'][:8], '%Y%m%d').timestamp(), axis=1).astype('int64')
data = data.sort_values(by=['date'])
data = data.reset_index(drop=True)
data["timestamp"] = data.index
```

```python
# Enum type must be string
data["id"] = data["id"].astype(str)
data["date"] = data.apply(lambda x: x['date'][:8], axis=1).astype('int64')
# data["zipcode"] = data["zipcode"].astype(str)
# data["waterfront"] = data["waterfront"].astype(str)
# data["view"] = data["view"].astype(str)
# data["condition"] = data["condition"].astype(str)


# Convert in craft ai python dictionnary structure
context_list = []
for index, row in data.iterrows():
    context_list.append({
        "timestamp" : row["timestamp"],
        "context" : {
#           "id": row["id"],
            "date": row["date"],
            "price": row["price"],
            "bedrooms": row["bedrooms"],
            "bathrooms": row["bathrooms"],
            "sqft_living": row["sqft_living"],
            "sqft_lot": row["sqft_lot"],
            "floors": row["floors"],
            "waterfront": row["waterfront"],
            "view": row["view"],
            "condition": row["condition"],
            "grade": row["grade"],
            "sqft_above": row["sqft_above"],
            "sqft_basement": row["sqft_basement"],
            "yr_built": row["yr_built"],
            "yr_renovated": row["yr_renovated"],
            "zipcode": row["zipcode"],
            "lat": row["lat"],
            "long": row["long"],
            "sqft_living15": row["sqft_living15"],
            "sqft_lot15": row["sqft_lot15"]
        }
    })

# Output
with open('./data/clean_data.json', 'w') as outfile:
    json.dump(context_list, outfile, indent = 3)
print("Saved to ./data/clean_data.json")
```
